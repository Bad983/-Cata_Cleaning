```python
import os
import pandas as pd
```

### Variabili Globali

```python
# PARAMETRI GLOBALI
root_folder = 'drive/MyDrive/Traduzione_Dantesca/Opere/Dante/'

# DATI
data_folder_name_originale = 'Originale'
filename_originale = 'ORIG_DETTO_AMORE.txt'

data_folder_name_traduzione = 'Traduzione'
filenamepath_traduzione_en = 'TRAD_1_DETTO_AMORE_EN.txt'
filenamepath_traduzione_it = 'TRAD_1_DETTO_AMORE_IT.txt'
filenamepath_traduzione_it_2 = 'TRAD_2_DETTO_AMORE.txt'

DATA_PATH_ORIGINAL = os.path.abspath(os.path.join(root_folder, data_folder_name_originale))
orig_filenamepath = os.path.abspath(os.path.join(DATA_PATH_ORIGINAL, filename_originale))

DATA_PATH_TRANSLATE = os.path.abspath(os.path.join(root_folder, data_folder_name_traduzione))
trad_filenamepath_en = os.path.abspath(os.path.join(DATA_PATH_TRANSLATE, filenamepath_traduzione_en))
trad_filenamepath_it = os.path.abspath(os.path.join(DATA_PATH_TRANSLATE, filenamepath_traduzione_it))
trad_filenamepath_it_2 = os.path.abspath(os.path.join(DATA_PATH_TRANSLATE, filenamepath_traduzione_it_2))

data_folder_out = 'drive/MyDrive/Traduzione_Dantesca/data'
DATA_PATH_OUT = os.path.abspath(os.path.join(data_folder_out, 'detto_amore.csv'))
```

```python
df_orig = pd.read_csv(orig_filenamepath, header=None, sep='\n', names=['Original'])
df_orig
```

```python
df_en = pd.read_csv(trad_filenamepath_en, header=None, sep='\n', names=['Translate_EN'])
df_en
```

```python
df_it = pd.read_csv(trad_filenamepath_it, header=None, sep='\n', names=['Translate_IT'])
df_it
```

```python
df_it_2 = pd.read_csv(trad_filenamepath_it_2, header=None, sep='\n', names=['Translate_2'])
df_it_2
```

```python
df = df_orig.join(df_en).join(df_it)
df
```

```python
df.to_csv(DATA_PATH_OUT, index=False)
```
