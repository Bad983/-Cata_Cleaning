```python
data_folder_in = 'drive/MyDrive/Traduzione_Dantesca/Data_Translate/data/'
data_folder_out = 'drive/MyDrive/Traduzione_Dantesca/Data_Translate/data_out/'

file_in = ['trad_divina_commedia_inferno_1_*.csv', 
           'trad_divina_commedia_inferno_2_*.csv',
           'trad_divina_commedia_inferno_3_*.csv',
           'trad_divina_commedia_inferno_4_*.csv',
           'trad_divina_commedia_inferno_5_*.csv',
           'trad_divina_commedia_purgatorio_1_*.csv', 
           'trad_divina_commedia_purgatorio_2_*.csv',
           'trad_divina_commedia_purgatorio_3_*.csv',
           'trad_divina_commedia_purgatorio_4_*.csv',
           'trad_divina_commedia_purgatorio_5_*.csv',
           'trad_divina_commedia_paradiso_1_*.csv', 
           'trad_divina_commedia_paradiso_2_*.csv',
           'trad_divina_commedia_paradiso_3_*.csv',
           'trad_divina_commedia_paradiso_4_*.csv',
           'trad_divina_commedia_paradiso_5_*.csv']

file_out = ['dataset_divina_commedia_inferno_1.csv',
            'dataset_divina_commedia_inferno_2.csv',
            'dataset_divina_commedia_inferno_3.csv',
            'dataset_divina_commedia_inferno_4.csv',
            'dataset_divina_commedia_inferno_5.csv',
            'dataset_divina_commedia_purgatorio_1.csv',
            'dataset_divina_commedia_purgatorio_2.csv',
            'dataset_divina_commedia_purgatorio_3.csv',
            'dataset_divina_commedia_purgatorio_4.csv',
            'dataset_divina_commedia_purgatorio_5.csv',
            'dataset_divina_commedia_paradiso_1.csv',
            'dataset_divina_commedia_paradiso_2.csv',
            'dataset_divina_commedia_paradiso_3.csv',
            'dataset_divina_commedia_paradiso_4.csv',
            'dataset_divina_commedia_paradiso_5.csv']
```

```python
def clean(text):
  text = re.sub('[^a-zA-Z?.!,]', ' ', text)
  text = re.sub('\([^)]*\)', '', text)
  return text
```

```python
for filaname_in, filename_out in zip(file_in, file_out):
  filenamepath_in = os.path.abspath(os.path.join(data_folder_in, filaname_in))
  filenamepath_out = os.path.abspath(os.path.join(data_folder_out, filename_out))

  df = pd.DataFrame(columns=['Filename', 'Original', 'Translate'])

  for filename in natsorted(glob.glob(filenamepath_in)):
    name = (filename.split('/')[-1])
   
    df_filename = pd.read_csv(filename)

    list_columns = [col for col in df_filename.columns if 'Translate_IT' in col] 
    
    for col in list_columns:
      df_temp = pd.DataFrame(columns=['Filename', 'Original', 'Translate'])
      df_temp['Filename'] = [name for _ in range(len(df_filename))]
      df_temp['Original'] = df_filename['Original']
      
      # df_temp['Original'] = [clean(text_orig) for text_orig in df_filename['Original']]
      df_temp['Translate'] = df_filename[col]
      df = pd.concat([df, df_temp])

  df = df.dropna()
  df = df.reset_index(drop=True)
  # df['Original'] =  [clean(text) for text in df['Original']]
  # df['Translate'] =  [clean(text) for text in df['Translate']]
  
  df.to_csv(filenamepath_out, index=False, encoding='utf-8')
```
